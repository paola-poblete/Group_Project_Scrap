```python
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
import os
import csv
import matplotlib.pyplot as plt

```

```python
query_url = 'https://api.ed.gov/data/mbk-labor-force-participation?'
api_key = 'api_key=bVIf4sU0IzjYfs473ZQEyzOjmZ5hkWeG6vdk5vPc'
per_page = 'per_page=' + '1258'
full_url = query_url + api_key + "&" + per_page

labour_data = requests.get(full_url).json()
#pprint(labour_data)
```

```python
labour_len = len(labour_data['resources'])
```

```python
#column lists
labour_percentage = []
labour_year = []
labour_race = []
labour_sex = []
labour_age = []

#for loop to create columns
for a in range(labour_len):
    labour_percentage.append(labour_data['resources'][a]['Percentage'])
    labour_year.append(int(labour_data['resources'][a]['Year']))
    labour_race.append(labour_data['resources'][a]['Race/ethnicity'])
    labour_sex.append(labour_data['resources'][a]['Sex'])
    labour_age.append(labour_data['resources'][a]['Age'])
    
#dataframe

labour_df = pd.DataFrame({
    "Year" : labour_year,
    "Race" : labour_race,
    "Sex" : labour_sex,
    "Age" : labour_age,
    "Percentage" : labour_percentage
})

#labour_df.head()
labour_df.sort_values(by=['Year', 'Race'], ascending=True)
```

```python
new_labour_df = labour_df.dropna(how='any')

new_labour_df
```

```python
#graph
#x axis = year
#y axis = percentage
#lines = race

labour_df_byrace = pd.DataFrame(labour_df.groupby(['Year', 'Race'])['Percentage'].sum())
labour_df_byrace.unstack(level=-1).plot(kind='line', legend=False)
plt.show()
```
